```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV, SGDClassifier
from sklearn.model_selection import learning_curve, validation_curve
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import make_scorer, roc_auc_score, accuracy_score
import warnings
import matplotlib.pyplot as plt
warnings.simplefilter("ignore")
```

```python
data = pd.read_csv("telecom_churn.csv")
data["Voice mail plan"] = data["Voice mail plan"].map({"No": 0, "Yes": 1})
data["International plan"] = (data["International plan"] == "Yes").astype(np.int32)
data["Churn"] = data["Churn"].astype("int")
data.drop("State", axis=1, inplace=True)
X = data.drop("Churn", axis=1).values
y = data["Churn"].values
```

```python
alphas = np.logspace(start=-2, stop=0, num=20)
sgd_logit = SGDClassifier(loss="log", n_jobs=-1, random_state=17, max_iter=5)
pipeline = Pipeline([("scaler", StandardScaler()),
                     ("poly", PolynomialFeatures(degree=2)),
                     ("sgd", sgd_logit)])
train_score, val_score = validation_curve(estimator=pipeline,
                                          X=X,
                                          y=y,
                                          param_name="sgd__alpha",
                                          param_range=alphas,
                                          scoring="roc_auc",
                                          cv=5)
```

```python
    plt.xlabel(r"$alphas$", fontsize=15)
    plt.ylabel(r"ROC AUC", fontsize=15)
    plt.legend(fontsize=15)
    plt.grid()
```

```python
def plot_validation_curve(parameter_list, scores, **kwargs):

    mu, std = scores.mean(axis=1), scores.std(axis=1)
    lines = plt.plot(parameter_list, mu, **kwargs)
    plt.fill_between(x=parameter_list,
                    y1=mu-std,
                    y2=mu+std,
                    edgecolor=None,
                    facecolor=lines[0].get_color(),
                    alpha=0.2,)
    plt.xlabel(r"$alphas$", fontsize=15)
    plt.ylabel(r"ROC AUC", fontsize=15)
    plt.legend(fontsize=15)
    plt.grid(True)
    plt.gcf().set_size_inches(w=10, h=6)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15);


plot_validation_curve(alphas, train_score, label="Training Scores")
plot_validation_curve(alphas, val_score, label="Val Scores")
```
