## 3. Sklearn Random Forest Feature Importance


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = iris["data"]
target = iris["target"]
```

```python
df = pd.DataFrame(data, columns=iris['feature_names'])
df.head(10)
```

Since our aim is just to demonstrate the sequence of steps in calculating feature importances we'll transform the target variable as for classifying Iris Virginica One-To-All.



```python
target = np.where((target==1) | (target==0), 0, 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=3, max_depth=3, random_state=17)
rfc.fit(data, target)
```

```python
rfc.base_estimator_
```

```python
rfc.estimators_
```

```python
(63/150)*(1-((61/63)**2 + (2/63)**2)) + (87/150)*(1-((28/87)**2 + (59/87)**2))
```

```python
1 - ((89/150)**2 + (61/150)**2)
```

```python
from sklearn import tree
for estimator in rfc.estimators_:
    plt.figure(figsize=(20, 12))
    tree.plot_tree(decision_tree=estimator, filled=True, feature_names=iris['feature_names'], class_names=['Y', 'N'], node_ids=True)
    plt.show()
```

## 4.Practical example

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('../datasets/hostel_factors.csv')
```

```python
df.head()
```

```python
features = {
    "f1": "Staff",
    "f2": "Hostel booking",
    "f3": "Check-in and check-out",
    "f4": "Room condition",
    "f5": "Shared kitchen condition",
    "f6": "Shared space condition",
    "f7": "Extra services",
    "f8": "General conditions & conveniences",
    "f9": "Value for money",
    "f10": "Customer Co-creation",
}
df.rename(columns=features, inplace=True)
```

```python
df
```

```python
X = df.drop(columns=['hostel'])
y = X.pop('rating')
```

```python
rfr = RandomForestRegressor(n_estimators=1000, max_features=10, random_state=0)
rfr.fit(X.values, y.values)
feature_importances_ = rfr.feature_importances_
```

```python
feature_importances_
```

```python
feature_indices = np.argsort(feature_importances_)[::-1]
```

```python
print("Feature Ranking:")
for i in feature_indices:
    print(X.columns[i] +": "+np.round(feature_importances_[i], 2).astype('str'))
```

```python
[str(i/11) for i in 1.0 + np.arange(len(feature_importances_)).astype('float')]
```

```python
[str(i) for i in np.linspace(0, 255, 11) / 255]
```

```python
X.columns
```

```python
feature_indices
```

```python
list(X.columns[feature_indices])
```

```python
plt.figure(figsize=(16, 12))
bars = plt.bar(x=range(len(feature_importances_)),
        height=feature_importances_[feature_indices],
        color=[str(i) for i in np.linspace(0, 255, 11) / 255])
plt.legend(bars, ['Staff',
 'Value for money',
 'Shared space condition',
 'Extra services',
 'Customer Co-creation',
 'General conditions & conveniences',
 'Shared kitchen condition',
 'Check-in and check-out',
 'Hostel booking',
 'Room condition'], fontsize=15)
```
